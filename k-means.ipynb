```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import os
import matplotlib

# 日本語フォント設定
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Yu Gothic', 'Meiryo', 'Hiragino Maru Gothic Pro', 'Hiragino Kaku Gothic Pro', 'MS Gothic']

# 代替方法（上記が動作しない場合）
try:
    import matplotlib.font_manager as fm
    font_path = 'C:/Windows/Fonts/msgothic.ttc'  # Windowsの場合のMSゴシック
    if os.path.exists(font_path):
        prop = fm.FontProperties(fname=font_path)
        matplotlib.rcParams['font.family'] = prop.get_name()
    # Macの場合
    mac_font_path = '/System/Library/Fonts/ヒラギノ角ゴシック W3.ttc'
    if os.path.exists(mac_font_path):
        prop = fm.FontProperties(fname=mac_font_path)
        matplotlib.rcParams['font.family'] = prop.get_name()
except Exception as e:
    print(f"フォント設定エラー: {e}")  # フォント設定に失敗しても処理を続行

def find_optimal_k(data, max_clusters=10):
    """エルボー法とシルエットスコアで最適なクラスター数を探索"""
    print("K-meansのエルボー法とシルエット分析を実行中...")
    
    # クラスター数ごとのイナーシャとシルエットスコアを計算
    inertia_values = []
    silhouette_values = []
    k_values = range(2, max_clusters + 1)
    
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(data)
        inertia_values.append(kmeans.inertia_)
        
        # シルエットスコアの計算（2クラスター以上でのみ計算可能）
        silhouette_avg = silhouette_score(data, labels)
        silhouette_values.append(silhouette_avg)
        print(f"クラスター数 {k}: イナーシャ = {kmeans.inertia_:.2f}, シルエットスコア = {silhouette_avg:.4f}")
    
    # 最適なクラスター数の特定（シルエットスコアが最大のもの）
    optimal_k = k_values[np.argmax(silhouette_values)]
    print(f"シルエットスコアに基づく最適クラスター数: {optimal_k}")
    
    # エルボー法とシルエットスコアのプロット
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # エルボー法（イナーシャのプロット）
    ax1.plot(k_values, inertia_values, 'bo-')
    ax1.set_xlabel('クラスター数', fontsize=12)
    ax1.set_ylabel('イナーシャ (Within-cluster sum of squares)', fontsize=12)
    ax1.set_title('K-means: エルボー法', fontsize=14)
    ax1.grid(True, alpha=0.3)
    
    # シルエットスコアのプロット
    ax2.plot(k_values, silhouette_values, 'go-')
    ax2.set_xlabel('クラスター数', fontsize=12)
    ax2.set_ylabel('シルエットスコア', fontsize=12)
    ax2.set_title('K-means: シルエット分析', fontsize=14)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    return optimal_k, fig, inertia_values, silhouette_values, k_values

def plot_clustered_data(data, clusters, point_size=50, alpha=0.8):
    """クラスターで色分けした可視化を行う関数"""
    
    # 図の作成
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # クラスターごとに色を設定
    colors = ['#1f77b4', '#7f7f7f', '#2ca02c', '#ff7f0e', '#9467bd']  # 青、グレー、緑、オレンジ、紫
    
    # クラスター数に応じて色を繰り返す
    n_clusters = len(np.unique(clusters))
    colors = colors[:n_clusters]  # 必要な色だけ取得
    
    # クラスターごとにポイントを描画
    scatter = ax.scatter(
        data[:, 0], 
        data[:, 1], 
        c=[colors[c] for c in clusters], 
        s=point_size,
        alpha=alpha,
        edgecolor='white',
        linewidth=0.5
    )
    
    # 軸の日本語ラベル表示対策
    plt.rcParams['axes.unicode_minus'] = False  # マイナス記号を正しく表示
    
    # 軸ラベルとタイトル
    ax.set_xlabel('特徴量1', fontsize=14)
    ax.set_ylabel('特徴量2', fontsize=14)
    ax.set_title('K-meansクラスタリングの結果', fontsize=16)
    ax.grid(True, alpha=0.3)
    
    # 凡例の追加
    for i in range(n_clusters):
        ax.scatter([], [], c=colors[i], label=f'クラスター {i}')
    ax.legend(prop={'family': 'sans-serif', 'size': 12})
    
    fig.tight_layout()
    return fig

# データの読み込みと前処理の関数
def load_and_preprocess_data(file_path, separator=','):
    """CSVファイルを読み込み、前処理を行う"""
    print(f"ファイルを読み込みます: {file_path}")
    
    # CSVファイルの読み込み
    df = pd.read_csv(file_path, sep=separator)
    print(f"データ読み込み完了: {df.shape[0]}行, {df.shape[1]}列")
    
    # 数値列のみを選択
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    print(f"特徴量として使用: {len(numeric_cols)}列")
    
    # 特徴量が空の場合はエラー
    if not numeric_cols:
        raise ValueError("数値特徴量が見つかりません")
    
    # 数値データを抽出
    X = df[numeric_cols].copy()
    X = X.fillna(X.mean())  # 欠損値を平均値で補完
    
    # データの標準化
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    return df, X_scaled, numeric_cols

# メイン処理
if __name__ == "__main__":
    # ファイルパスの設定
    path = "G:/共有ドライブ/GAP_長寿研/user/iwamoto/視線の動きの俊敏さ/data"
    file_path = f"{path}/features_test_nolabel.csv"
    
    # データの読み込みと前処理
    df, X_scaled, numeric_cols = load_and_preprocess_data(file_path)
    
    # データのサイズを確認
    print(f"データサイズ: {X_scaled.shape}")
    
    # K-meansのエルボー法とシルエット分析
    optimal_k, kmeans_fig, _, _, _ = find_optimal_k(X_scaled, max_clusters=8)
    
    # ユーザーが指定したクラスター数または最適なクラスター数を使用
    n_clusters = 3  # print(f"クラスター数を{n_clusters}に設定しました")
    
    # クラスタリングの実行
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X_scaled)
    
    # クラスター情報をデータフレームに追加
    df['cluster'] = clusters
    
    # クラスターごとのポイント数をカウント
    cluster_counts = np.bincount(clusters)
    for i, count in enumerate(cluster_counts):
        print(f"クラスター {i}: {count} ポイント")

    # 結果を保存するディレクトリ
    result_path = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ\result"
    os.makedirs(result_path, exist_ok=True)
    
    # K-meansのエルボー法グラフを保存
    kmeans_plot_path = os.path.join(result_path, "kmeans_elbow_method.png")
    kmeans_fig.savefig(kmeans_plot_path, dpi=300, bbox_inches='tight')
    print(f"K-meansのエルボー法グラフを保存しました: {kmeans_plot_path}")
    
    # 次元削減が必要な場合は、以下のようにして2次元データを作成してから可視化
    # 簡略化のために、最初の2つの特徴量だけを使用した例
    simplified_data = X_scaled[:, :2]  # 最初の2つの特徴量を使用
    
    # クラスタリング結果の可視化
    cluster_fig = plot_clustered_data(simplified_data, clusters)
    
    # クラスタリング結果のプロットを保存
    cluster_plot_path = os.path.join(result_path, "kmeans_clustering.png")
    cluster_fig.savefig(cluster_plot_path, dpi=300, bbox_inches='tight')
    print(f"クラスタリング結果のプロットを保存しました: {cluster_plot_path}")
    
    # データを保存
    csv_path = os.path.join(result_path, "clustered_data.csv")
    df.to_csv(csv_path, index=False)
    print(f"クラスター情報付きデータを保存しました: {csv_path}")
    
    # プロットを表示
    plt.show()
```
